```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.metrics.pairwise import cosine_similarity
from src.models import db, Laptop, Cpu, Gpu, Windows as Wind, Application as App
from src import app
```

### GET LAPTOP FROM DATABASE

```python
laptops_full = pd.read_csv("dev-data/data.test.csv")
laptops_full["totalStorage"] = laptops_full["SSD"] + laptops_full["HDD"]

laptops_full.head()
```

### REMOVE UNUSED PROPERTY

```python
imp_laptop = laptops_full.drop(columns=["no", "name", "HDD", "SSD", 'GPU',"CPU","Windows", "Edition"])
imp_laptop.head()
```

### NORMALIZE

```python
normalized = imp_laptop.apply(zscore)
normalized.head()
```

### FIND COSINE SIMILARITY

```python
similarity = cosine_similarity(normalized, normalized.values[1].reshape(1, -1))
laptops_full["similarity"] = similarity

laptops_full.sort_values(by="similarity", ascending=False).head(6)
```
